# 参数指定

```python
# cand_train, cand_test, query_train, query_test = train_test_split(Cand, Query, random_state=1, train_size=0.7)    

cand_test_data = transformed_trajectory[0:10]
query_test_data = transformed_trajectory[10:20]

minLen =10
error_dict_for_SBLCS = [0.15, 0.15]
```

# 生成答案

```python
length = len(cand_test_data)
elist = [i for i in range(length)]
```

```python
SUBSIM_TEST = []
SUBSIM_RANK = []
for i in elist:
    print('process', i)
    subsim, subtraj, subset = ExactS(cand_test_data[i], query_test_data[i], minLen)
    print('lens cand:', len(cand_test_data[i]), 'lens sub:', len(subset))
    SUBSIM_TEST.append(subsim)
    subsort = sorted(subset.items(), key=lambda d: d[1])
    SUBSIM_RANK.append([j[0] for j in subsort])
```

```python
pickle.dump(SUBSIM_TEST, open('SUBSIM_TEST', 'wb'), protocol=2)
pickle.dump(SUBSIM_RANK, open('SUBSIM_RANK', 'wb'), protocol=2)
```

```python
SUBSIM_TEST = pickle.load(open( 'SUBSIM_TEST', 'rb'), encoding='bytes')
SUBSIM_RANK = pickle.load(open( 'SUBSIM_RANK', 'rb'), encoding='bytes')
```

```python
for oo in ['SBLCS']: 
    print('algorithm:', oo)
    opt = oo
    eva = []
    rank = []
    relative_rank = []
    for i in elist:
        print('steps:', i, len(cand_test_data[i]), len(SUBSIM_RANK[i]))
        total = (len(cand_test_data[i]) * (len(cand_test_data[i]) + 1) / 2) * (len(query_test_data[i]) * (len(query_test_data[i]) + 1) / 2)
        if opt == 'SBLCS':
            error_dict = error_dict_for_SBLCS
            _, ap_subsim, ap_subtraj, subset  = SBLCS(cand_test_data[i], query_test_data[i], minLen, error_dict) 

        if SUBSIM_TEST[i] != 0:
            eva.append(ap_subsim/SUBSIM_TEST[i])
        elif ap_subsim == SUBSIM_TEST[i]:
            eva.append(1.0)
        if tuple(ap_subtraj) in SUBSIM_RANK[i]:
            t = SUBSIM_RANK[i].index(tuple(ap_subtraj)) + 1
            rank.append(t)
            relative_rank.append(t*1.0/total)
        if i + 1 == length:
            print('average competive ratio:', sum(eva)/len(eva))
            print('mean rank:', sum(rank)/len(rank))
            print('relative rank:', sum(relative_rank)/len(relative_rank))
            break
```
